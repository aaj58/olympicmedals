```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 250)
```

```python
winter = pd.read_csv(r'C:\Users\ashle\OneDrive\Documents\Data Fun\Olympics\winter.csv')
summer = pd.read_csv(r'C:\Users\ashle\OneDrive\Documents\Data Fun\Olympics\summer.csv')
```

```python
wmedals_per_country=winter.groupby(['Country', 'Medal'])['Year'].count().reset_index().sort_values(by='Country',ascending=True)
wmedals_per_country=wmedals_per_country.pivot(index='Country', columns='Medal', values='Year').fillna(0)
wmedals_per_country['Total'] = wmedals_per_country.sum(axis=1)
wmedals_per_country = wmedals_per_country.sort_values(by='Total',ascending=False)
wmedals_per_country['Cumulative_Sum_Gold'] = wmedals_per_country.Gold.cumsum()
wmedals_per_country['Cumulative_Percent_Gold'] = 100*wmedals_per_country.Cumulative_Sum_Gold/wmedals_per_country.Gold.sum()
wmedals_per_country['Cumulative_Sum_All'] = wmedals_per_country.Total.cumsum()
wmedals_per_country['Cumulative_Percent_All'] = 100*wmedals_per_country.Cumulative_Sum_All/wmedals_per_country.Total.sum()
wmedals_per_country.round(2)

smedals_per_country=summer.groupby(['Country', 'Medal'])['Year'].count().reset_index().sort_values(by='Country',ascending=True)
smedals_per_country=smedals_per_country.pivot(index='Country', columns='Medal', values='Year').fillna(0)
smedals_per_country['Total'] = smedals_per_country.sum(axis=1)
smedals_per_country = smedals_per_country.sort_values(by='Total',ascending=False)
smedals_per_country['Cumulative_Sum_Gold'] = smedals_per_country.Gold.cumsum()
smedals_per_country['Cumulative_Percent_Gold'] = 100*smedals_per_country.Cumulative_Sum_Gold/smedals_per_country.Gold.sum()
smedals_per_country['Cumulative_Sum_All'] = smedals_per_country.Total.cumsum()
smedals_per_country['Cumulative_Percent_All'] = 100*smedals_per_country.Cumulative_Sum_All/smedals_per_country.Total.sum()
smedals_per_country.round(2)

wintercountries=float(len(winter.Country.unique()))
summercountries=float(len(summer.Country.unique()))
print 'Total Countries Participated in Winter Olympics '+ str(wintercountries)
print 'Total Countries Participated in Summer Olympics '+ str(summercountries)
```

```python
print '80% of Medals Won by ' + str(13/wintercountries*100) + '% of countries.'
wtop_country=wmedals_per_country.sort_values(by='Total',ascending=False)[:14]
wmedals=wtop_country[['Gold', 'Silver', 'Bronze']]
wmedals.plot.bar(width=0.75,stacked=True,color=['#D4AF37','#C0C0C0','#CD7F32'])
# wtop_country['Cumulative_Percent_All'].plot.line()
plt.title('Medals from Top 80% Countries (Winter)')
plt.xlabel('Country')
plt.ylabel('Count')
plt.savefig('Winter_Countries.png')
plt.show()
```
